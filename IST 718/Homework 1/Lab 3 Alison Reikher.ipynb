```python
#This block of code was added because of issues with importing the statsmodels
#This is what was suggested on the internet and it works
import  scipy.signal.signaltools

def _centered(arr, newsize):
    # Return the center newsize portion of the array.
    newsize = np.asarray(newsize)
    currsize = np.array(arr.shape)
    startind = (currsize - newsize) // 2
    endind = startind + newsize
    myslice = [slice(startind[k], endind[k]) for k in range(len(endind))]
    return arr[tuple(myslice)]

scipy.signal.signaltools._centered = _centered
```

```python
from scipy.stats import uniform  # for training-and-test split
import statsmodels.api as sm  # statistical models (including regression)
import statsmodels.formula.api as smf  # R-like model specification
```

```python
import numpy as np  # arrays and math functions
# employ training-and-test regimen for model validation
np.random.seed(1234)
data['runiform'] = uniform.rvs(loc = 0, scale = 1, size = len(data))
data_train = data[data['runiform'] >= 0.33]
data_test = data[data['runiform'] < 0.33]
# check training data frame
print('\ncouch_train data frame (rows, columns): ',data_train.shape)
print(data_train.head())
# check test data frame
print('\ncouch_test data frame (rows, columns): ',data_test.shape)
print(data_test.head())
```

```python

# MODEL

# specify a simple model with TotalPay as a function of Wins
my_model_simple1 = str('TotalPay ~ Wins')

# fit the model to the training set
train_model_fit1 = smf.ols(my_model_simple1, data = data_train).fit()
# summary of model fit to the training set
print(train_model_fit1.summary())
# training set predictions from the model fit to the training set
data_train['predict_wins'] = train_model_fit1.fittedvalues

# test set predictions from the model fit to the training set
data_test['predict_wins'] = train_model_fit1.predict(data_test)
```

```python
#creates colorful correlation matrix
corr = data.corr()

# Generate a mask for the upper triangle
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True


# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
plt.show()
```
